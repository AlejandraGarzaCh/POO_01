```python
#Vamos a comparar estadísticamente una lista de tickers
#Vamos a descargar tres tickers
fecha_inicial = "2021-12-31"
fecha_final = "2023-10-05"
```

```python
femsa = yf.download("FEMSAUBD.MX", fecha_inicial, fecha_final)
herdez = yf.download("HERDEZ.MX", fecha_inicial, fecha_final)
cuervo = yf.download("CUERVO.MX", fecha_inicial, fecha_final)
alsea = yf.download("ALSEA.MX", fecha_inicial, fecha_final)
ac = yf.download("AC.MX", fecha_inicial, fecha_final)
```

```python
#Lista de tickers
tickers = ["FEMSAUBD.MX", "HERDEZ.MX", "CUERVO.MX", "ALSEA.MX", "AC.MX" ]
datos = yf.download(tickers, fecha_inicial, fecha_final)["Close"]
datos
```

```python
#Gráfica
datos.describe()
```

```python
#Para calcular la varianza de los precios de cada ticker
varianza = datos.std()**2
varianza
```

```python
#Otra manera de calcular la varianza de los precios
varianza = datos.var()
varianza
```

```python
#Para sacar los rendimientos
rendimientos = datos.pct_change()
rendimientos
```

```python
var = rendimientos.var()*100
var
```

```python
rendimientos.mean()
```

```python
rendimientos.describe()
```

```python
# Matriz de covarianza
covarianza = rendimientos.cov()
covarianza
```

```python
# Dropna es para quitar el dato de NA que te pone en el primer renglon columna rendimientos
herdez ["Rendimientos"] = (herdez["Close"]/ herdez["Close"].shift(1))-1
herdez = herdez [["Close", "Rendimientos"]].dropna()
herdez
```
